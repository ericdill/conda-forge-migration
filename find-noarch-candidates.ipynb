```python
linux['ext'] = linux['ext'].apply(eval)
```

```python
# Define some common binary extensiosn

bin_extensions = set(['.so', '.pyd', '.dylib', '.exe', '.dll', '.lib', '.a'])
```

```python
categorized_extensions = {
    'numeric': ['.%d' % num for num in range(10)],
    'web': ['.css', '.css_t', '.html', '.gif', '.js', '.png'],
    'jupyter': ['.ipynb'],
    'conda': ['.rendered', '.template', '.yaml', '.bat', '.sh',
             '.tmpl', '.yml'],
    'other': ['', '.py', '.pyc', '.json', '.json', '.egg-info', 
              '.txt', '.pc', '.rst', '.csv', '.ogg', '.test',
              '.conf', '.xsd', '.xml', '.ico', '.map', '.cmd',
              '.svg', '.manifest', '.info', '.bash', '.md',
              '.dat']
}
allowed_extensions = set()
for category, extensions in categorized_extensions.items():
    allowed_extensions.update(extensions)
```

```python
linux['compiled'] = linux['ext'].apply(lambda extensions: any(ext in extensions for ext in bin_extensions))
linux['pure'] = linux['ext'].apply(lambda extensions: extensions <= allowed_extensions)
linux['neither'] = ~linux['pure'] & ~linux['compiled']
```

```python
total = 0
for col in ['compiled', 'pure', 'neither']:
    subset = linux[linux[col] == True]
    print('%s: %s' % (col, len(subset)))
    total += len(subset)
print(total, len(linux))
```

```python
neither = (linux[linux['neither'] == True]).copy()
```
